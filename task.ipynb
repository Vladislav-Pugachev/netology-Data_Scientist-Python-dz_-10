# Задание №1

```python
import pandas as pd
```

```python
rainting= pd.read_csv('ml-25m/ratings.csv')
movies = pd.read_csv('ml-25m/movies.csv')
#Максимальный регйинг
raiting_5 = rainting['rating'].max()
#Список фильмов с максимальным рейтингом
movies_with_max_raiting =rainting[rainting.rating==raiting_5]
#ID фильма с максимальным рейтингом
id_movies_with_max_raiting=movies_with_max_raiting['movieId'].value_counts().index[0]
#Название фильма с максимальным рейтингом по ID
name_movies_with_max_raiting = movies[movies.movieId==id_movies_with_max_raiting]
print(name_movies_with_max_raiting['title'].iloc[0])
```

# Задание  №2

```python
data = pd.read_csv('power.csv')
#Фильтр по странам
filtered_countries = data[ (data['country']=='Latvia') | (data['country']=='Lithuania') | (data['country']=='Estonia')]

#Фильтр по категоряим потребления
filtered_countries_categ= filtered_countries[(filtered_countries['category']==float('4')) | (filtered_countries['category']==float('12')) | (filtered_countries['category']==float('21'))]
#Фильтр по годам
filtered_countries_categ_year=filtered_countries_categ[ (filtered_countries_categ['year']>=int('2005')) & (filtered_countries_categ['year']<=int('2010'))]

#Сумма потреления
filtered_countries_categ_year_summ_quan=filtered_countries_categ_year[filtered_countries_categ_year['quantity']>float('0.0')]['quantity']
filtered_countries_categ_year_summ_quan.sum()
```
